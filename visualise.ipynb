```python
import numpy as np
import pandas as pd

data = pd.read_csv('heart.csv')
```

```python
import seaborn as sns
sns.scatterplot(x=data['Age'],y=data['Cholesterol'],hue=data['Sex'])
```

```python
sns.jointplot(x=data['Age'],y=data['Cholesterol'], kind='kde')
```

```python
sns.histplot(x=data['Cholesterol'])
```

```python
sns.kdeplot(x=data['Cholesterol'],shade=True)
```

```python
sns.lineplot(data=data)
```

```python
sns.lineplot(data=data['Cholesterol'], label='Cholesterol')
sns.lineplot(data=data['Age'], label='Age')
```
